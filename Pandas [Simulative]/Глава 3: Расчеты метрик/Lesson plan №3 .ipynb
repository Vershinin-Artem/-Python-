# 3.1 ABC-анализ

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns',40)
pd.set_option('display.float_format', '{:2f}'.format)
```

```python
df = pd.read_csv('data.csv', encoding = '1251')
df.head()
```

```python
group_df = df.groupby(['DR_NDrugs']).agg({'DR_Kol' : sum})
group_df
```

#### Оределяем долю каждого товара в итоговой сумме

```python
group_df = (group_df['DR_Kol'] / sum(group_df['DR_Kol'])).sort_values(ascending=False)
group_df
```

#### Считаем сумму продаж накопительным итогом

```python
group_df = group_df.cumsum().reset_index()
group_df
```

#### Определим группу

```python
group_df['abc'] = np.where(group_df['DR_Kol'] < 0.8, 'A', np.where(group_df['DR_Kol'] < 0.95, 'B', 'C'))
group_df
```

#### Выполним многомерный ABC анализ

```python
group_mm_df = df.groupby(['DR_NDrugs']).agg({'DR_Kol' : sum, 'DR_CRoz' : sum})
group_mm_df['rel_col'] = (group_mm_df['DR_Kol'] / sum(group_mm_df['DR_Kol'])).sort_values(ascending=False)
group_mm_df['cumsum_col'] = group_mm_df['rel_col'].cumsum()
group_mm_df['abc_col'] = np.where(group_mm_df['cumsum_col'] < 0.8, 'A', np.where(group_mm_df['cumsum_col'] < 0.95, 'B', 'C'))

group_mm_df['rel_cRoz'] = (group_mm_df['DR_CRoz'] / sum(group_mm_df['DR_CRoz'])).sort_values(ascending=False)
group_mm_df['cumsum_cRoz'] = group_mm_df['rel_cRoz'].cumsum()
group_mm_df['abc_cRoz'] = np.where(group_mm_df['cumsum_cRoz'] < 0.8, 'A', np.where(group_mm_df['cumsum_cRoz'] < 0.95, 'B', 'C'))

group_mm_df
```

#### Функция для автоматизации многомерного анализа 

```python
df1 = df[['DR_NDrugs', 'DR_Kol', 'DR_CRoz']]
df1.columns
```

```python
def perform_abc(df, index):
    """
    На вход подается датафрейм пандас...
    """
    cols = list(df.columns)
    cols.remove(index)
    groupped_df = df.groupby(index).agg({col: 'sum' for col in cols})
    for col in cols:
        groupped_df[f'rel_{col}'] = groupped_df[col] / sum(groupped_df[col])
        groupped_df = groupped_df.sort_values(f'rel_{col}', ascending=False)
        groupped_df[f'cumsum_{col}'] = groupped_df[f'rel_{col}'].cumsum()
        groupped_df[f'abc_{col}'] = np.where(groupped_df[f'cumsum_{col}'] < 0.8, 'A', np.where(groupped_df[f'cumsum_{col}'] < 0.95, 'B', 'C'))
    return groupped_df[cols + [f'abc_{col}' for col in cols]]
        
```

```python
perform_abc(df1, index='DR_NDrugs')
```

# 3.2 Динамика продаж

```python
data = pd.read_csv('data1.csv', encoding = '1251')
data.head()
```

#### Посмотрим на даты за которые у нас есть данные 

```python
data.DR_Dat.unique()
```

#### Посчитаем сумму продаж за каждый день

```python
data = data.groupby(['DR_Dat']).apply(lambda x: pd.Series({'revenue' : sum(x['DR_Kol'] * x['DR_CRoz'] - x['DR_SDisc'])})).reset_index()
data
```

```python
data['revenue_rol'] = data['revenue'].rolling(2).apply(lambda x: (x.iloc[1] - x.iloc[0]) / x.iloc[0]) # при омощи метода apply
data
```

```python
data['reveue_shifted'] = data['revenue'].shift(1) # при помощи метода shift
data['revenue_din_shift'] = (data['revenue'] - data['reveue_shifted'] ) / data['reveue_shifted'] 
data
```

```python
data['revenue_pct_change'] = data['revenue'].pct_change() # считаем динамику при помощи метода pct_change
data
```

# 3.3 XYZ-анализ

```python
xz_df = pd.read_csv('data1.csv', encoding = '1251')
xz_df.head()
```

```python
xz_df = xz_df.groupby(['DR_Dat', 'DR_NDrugs']).agg({'DR_Kol' : sum})
xz_df
```

```python
xz_df_1 = xz_df.reset_index().groupby('DR_NDrugs').agg({'DR_Dat': 'count'}).reset_index()
xyz_names = list(xz_df_1[xz_df_1['DR_Dat'] > 1]['DR_NDrugs'])
xyz_names
```

```python
xz_df = xz_df.reset_index()
xz_df = xz_df[xz_df['DR_NDrugs'].isin(xyz_names)]
xz_df = xz_df.groupby('DR_NDrugs')[['DR_Kol']].apply(lambda x: x.std()/x.mean())
```

```python
xz_df['XYZ'] = np.where(xz_df['DR_Kol'] < 0.1, 'X', np.where(xz_df['DR_Kol'] < 0.25, 'Y', 'Z'))
```
